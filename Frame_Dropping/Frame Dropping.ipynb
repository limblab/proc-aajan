```python
import glob
import numpy as np
```

```python
path = '/Volumes/L_MillerLab/data/Pop_18E3/Videos'
dates = ['20210712','20210814','20210816','20211105','20211108','20211111','20220203','20220210','20220224','20220309','20220405']
```

```python
file_list = []
for date in dates:
    x = glob.glob(f"{path}/{date}/*ata*/*logfile.txt")
    for i in x:
        file_list.append(i)
```

```python
%%time
diffs_list = []
for f in file_list:
    diffs = []
    x = np.genfromtxt(f)
    for i in range(len(x)-1):
        diffs.append(x[i+1]-x[i])
    diffs = np.array(diffs)
    frames = np.rint(diffs/0.033)
    diffs_list.append((np.where(frames>1)[0], frames[frames>1]))
```

```python
diffs_list
```

```python
test_logfile = file_list[2]
test_diffs = diffs_list[2]
```

```python
test_video = test_logfile.replace("_logfile.txt", ".avi")
```

```python
import cv2
vidcap = cv2.VideoCapture(test_video)

l = []
count = 0
while vidcap.isOpened():
    ret, img = vidcap.read()
    if ret:
        l.append(img)
        print('Read a new frame: ', str(count), str(ret))
    else:
        print('Reading failed!')
        break
    
#     cv2.imshow("camout", img)
    count+=1
    cv2.waitKey(1)
vidcap.release()
```
